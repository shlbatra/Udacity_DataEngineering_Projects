```python
# Get port locations from SAS text file
with open("./I94_SAS_Labels_Descriptions.SAS") as f:
    content = f.readlines()
```

```python
#Get Port code and location data
content = [x.strip() for x in content]
port = content[302:962]
port_data = [p.split("=") for p in port]
port_code = [x[0].replace("'","").strip() for x in port_data]
port_location = [x[1].replace("'","").strip() for x in port_data]
port_cityname = [x.split(",")[0] for x in port_location]
port_statename = [x.split(",")[-1] for x in port_location]
df_port = pd.DataFrame({"port_code" : port_code, "port_city": port_cityname, "port_state": port_statename})
df_port.head(5)
```

```python
#Remove Missing temperature values
df_temp_us.dropna(inplace=True)
df_temp_us["Year"] = df_temp_us["dt"].apply(lambda x:x[0:4])
df_temp_us["Month"] = df_temp_us["dt"].apply(lambda x:x[5:7])
df_temp_us.shape
```

```python
df_temp_us.head(5)
```

```python
# Remove Missing airport codes
df_airport.dropna(subset=['iata_code'],inplace=True)
df_airport.head(5)
```

```python
#Get the data to City Level
df_demo_city = df_demo[['City', 'State', 'Median Age', 'Male Population', 'Female Population',\
       'Total Population', 'Number of Veterans', 'Foreign-born','Average Household Size'
         , 'State Code']].drop_duplicates()
df_demo_city.head(5)
```

```python
#Remove columns with missing values in i95 immigration dataset
df_i94_na = pd.DataFrame({"column" : df_i94.columns, "Missing Rows": df_i94.isna().sum(), "Missing Rows %": round(df_i94.isna().sum()/df_i94.shape[0],2)})
```

```python
df_i94_na.sort_values(by="Missing Rows %",ascending=False).head(5)
```

```python
#Drop columns with high %age of missing values
df_i94_clean = df_i94.drop(['occup', 'entdepu','insnum','visapost','gender'], axis=1)
```

```python
#Drop columns not relevant
df_i94_clean.drop(['cicid','dtadfile', 'entdepa', 'entdepd','dtaddto'], axis=1,inplace=True)
```

```python
#airport dimension
df_dim_airport = df_airport.merge(df_port, left_on="iata_code", right_on="port_code")
df_dim_airport.drop(columns=["port_code"], inplace=True)
df_dim_airport = df_dim_airport[["iata_code", "name", "type", "local_code", "coordinates"\
                                 , "port_city", "elevation_ft", "continent", "iso_country", "iso_region", "municipality", "gps_code"]]
df_dim_airport.head(5)
```

```python
#demographics dimension
df_dim_demographics = df_demo_city.merge(df_port, left_on=df_demo_city["City"].str.lower(), right_on=df_port["port_city"].str.lower())
df_dim_demographics = df_dim_demographics[["City","State","Median Age","Male Population","Female Population",\
                                         "Total Population","Number of Veterans","Foreign-born","Average Household Size",\
                                         "State Code","port_code","port_city","port_state"]]
df_dim_demographics.head(10)
```

```python
#temperature dimension
df_dim_temperature = df_temp_us.merge(df_port, left_on=df_temp_us["City"].str.lower(), right_on=df_port["port_city"].str.lower())
df_dim_temperature = df_dim_temperature[["dt","AverageTemperature","AverageTemperatureUncertainty","City","Country","Latitude",
                                           "Longitude","port_code","port_state","Year","Month"]]
df_dim_temperature.head(10)
```

```python
#Fact i94 table
df_fact_i94 = df_i94_clean.merge(df_port, left_on=df_i94_clean["i94port"].str.lower(), right_on=df_port["port_code"].str.lower(),how="left")
df_fact_i94 = df_fact_i94[['i94yr', 'i94mon', 'i94cit', 'i94res', 'i94port', 'arrdate',
       'i94mode', 'i94addr', 'depdate', 'i94bir', 'i94visa', 'count',
       'matflag', 'biryear', 'airline', 'admnum', 'fltno', 'visatype',
       'port_code', 'port_city', 'port_state']]
df_fact_i94.head(3)
```

```python
.head(5)
```

```python
# Use case of Fact & Dimension data model - Number of people arriving in a city compared to its population
df_usecase = df_fact_i94.groupby(['port_code','port_city'])['count'].sum().reset_index().merge(
                        df_dim_demographics[['port_code','Total Population']].drop_duplicates(),
                        left_on="port_code", right_on="port_code")
df_usecase.head(10)
```

```python
## Data Quality Check
def quality_check(df,df_name):
    '''
    Input: DataFrame, DataFrame name
    Output: Data quality check for empty dataframe
    '''
    
    result = df.shape[0]
    if result == 0:
        print("Data quality check failed for {} with zero records".format(df_name))
    else:
        print("Data quality check passed for {} with {} records".format(df_name, result))
```
